```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('/home/chinmay/Documents/Internship/TASK2/Mall_Customers.csv')
data.head()
```

```python
fig, ax = plt.subplots(2, figsize=(5,10))
sns.scatterplot(ax=ax[0], data=data, x='Age', y='Spending Score (1-100)', hue='Gender')
sns.scatterplot(ax=ax[1], data=data, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender')
plt.show()
```

```python
km = KMeans(n_clusters=5)
data['cluster'] = km.fit_predict(data[['Annual Income (k$)', 'Spending Score (1-100)']])
data.head()
```

```python
km.cluster_centers_
```

```python
km.inertia_
```

```python
data['cluster'].unique()
```

```python
data0 = data[data['cluster'] == 0][['Annual Income (k$)', 'Spending Score (1-100)', 'cluster']]
data1 = data[data['cluster'] == 1][['Annual Income (k$)', 'Spending Score (1-100)', 'cluster']]
data2 = data[data['cluster'] == 2][['Annual Income (k$)', 'Spending Score (1-100)', 'cluster']]
data3 = data[data['cluster'] == 3][['Annual Income (k$)', 'Spending Score (1-100)', 'cluster']]
data4 = data[data['cluster'] == 4][['Annual Income (k$)', 'Spending Score (1-100)', 'cluster']]
```

```python
sns.scatterplot(data=data0,x='Annual Income (k$)',y='Spending Score (1-100)')
sns.scatterplot(data=data1,x='Annual Income (k$)',y='Spending Score (1-100)')
sns.scatterplot(data=data2,x='Annual Income (k$)',y='Spending Score (1-100)')
sns.scatterplot(data=data3,x='Annual Income (k$)',y='Spending Score (1-100)')
sns.scatterplot(data=data4,x='Annual Income (k$)',y='Spending Score (1-100)')
plt.show()
```

```python
sse = []
for k in range(1,20):
    kmk = KMeans(n_clusters=k)
    kmk.fit(data[['Annual Income (k$)', 'Spending Score (1-100)']])
    sse.append(kmk.inertia_)

sse
```

```python
sns.lineplot(data=sse)
sns.scatterplot(data=sse)
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Sum of Squared Error')
plt.title('Elbow Chart')
plt.show()
```
